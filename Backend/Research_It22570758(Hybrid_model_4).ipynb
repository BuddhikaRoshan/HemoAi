```python
# ========================
# Mount Drive
# ========================
from google.colab import drive
drive.mount('/content/drive')

# ========================
# Imports
# ========================
import os, glob, pickle
import cv2
import numpy as np
import tensorflow as tf
from tensorflow.keras import layers, models, regularizers
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.metrics import classification_report, accuracy_score
from sklearn.utils.class_weight import compute_class_weight
import matplotlib.pyplot as plt

# ========================
# Config
# ========================
IMG_SIZE = 128
BATCH_SIZE = 16
MAX_RBC_PER_IMAGE = 25
EPOCHS = 30
SEED = 42
np.random.seed(SEED)
tf.random.set_seed(SEED)

# ========================
# Handcrafted Feature Functions
# ========================
def target_cell_score(gray):
    h, w = gray.shape
    cx, cy = w//2, h//2
    Y, X = np.ogrid[:h, :w]
    d = np.sqrt((X - cx)**2 + (Y - cy)**2)
    r = np.max(d)
    c = gray[d < r*0.25]
    m = gray[(d >= r*0.25) & (d < r*0.5)]
    o = gray[(d >= r*0.5) & (d < r*0.75)]
    if len(c)==0 or len(m)==0 or len(o)==0:
        return 0
    return (np.mean(m)-np.mean(c)) + (np.mean(m)-np.mean(o))

def blue_ratio(rbc):
    b, g, r = cv2.split(rbc)
    return np.mean(b) / (np.mean(r)+1e-5)

def compute_rbc_features(rbc):
    gray = cv2.cvtColor(rbc, cv2.COLOR_BGR2GRAY)
    h, w = gray.shape
    return [
        target_cell_score(gray),
        blue_ratio(rbc),
        abs(h - w)/max(h, w),
        h*w
    ]

# ========================
# RBC Extraction Function
# ========================
def extract_rbcs(image):
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    blur = cv2.GaussianBlur(gray, (5,5), 0)
    _, th = cv2.threshold(blur, 0, 255, cv2.THRESH_BINARY_INV + cv2.THRESH_OTSU)
    contours, _ = cv2.findContours(th, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

    cells, feats = [], []
    for c in contours[:MAX_RBC_PER_IMAGE]:
        if cv2.contourArea(c) < 120:
            continue
        x, y, w, h = cv2.boundingRect(c)
        rbc = cv2.resize(image[y:y+h, x:x+w], (IMG_SIZE, IMG_SIZE))
        cells.append(rbc)
        feats.append(compute_rbc_features(rbc))
    return np.array(cells), np.array(feats)

# ========================
# Load Images (Image-level split)
# ========================
healthy_path  = '/content/drive/MyDrive/withhealthyrbc/healthy'
affected_path = '/content/drive/MyDrive/withhealthyrbc/affected'

healthy_imgs = glob.glob(os.path.join(healthy_path, '*'))
affected_imgs = glob.glob(os.path.join(affected_path, '*'))

all_images = healthy_imgs + affected_imgs
all_labels = ['healthy']*len(healthy_imgs) + ['thalassemia']*len(affected_imgs)

# Shuffle dataset
idx = np.arange(len(all_images))
np.random.shuffle(idx)
all_images = np.array(all_images)[idx]
all_labels = np.array(all_labels)[idx]

# Split images 70% train, 15% val, 15% test
train_end = int(0.7*len(all_images))
val_end   = int(0.85*len(all_images))

train_imgs, val_imgs, test_imgs = all_images[:train_end], all_images[train_end:val_end], all_images[val_end:]
train_labels, val_labels, test_labels = all_labels[:train_end], all_labels[train_end:val_end], all_labels[val_end:]

# ========================
# Extract RBCs per split
# ========================
def build_dataset(img_paths, labels):
    X, f, y = [], [], []
    for img_path, label in zip(img_paths, labels):
        img = cv2.imread(img_path)
        if img is None: continue
        rbcs, feats = extract_rbcs(img)
        if len(rbcs)==0: continue
        X.extend(rbcs)
        f.extend(feats)
        y.extend([label]*len(rbcs))
    return np.array(X), np.array(f), np.array(y)

X_train, f_train, y_train = build_dataset(train_imgs, train_labels)
X_val,   f_val,   y_val   = build_dataset(val_imgs, val_labels)
X_test,  f_test,  y_test  = build_dataset(test_imgs, test_labels)

print("Train samples:", len(X_train))
print("Val samples:  ", len(X_val))
print("Test samples: ", len(X_test))

# ========================
# Encode & Scale
# ========================
le = LabelEncoder()
y_train_enc = le.fit_transform(y_train)
y_val_enc   = le.transform(y_val)
y_test_enc  = le.transform(y_test)

scaler = StandardScaler()
f_train_scaled = scaler.fit_transform(f_train)
f_val_scaled   = scaler.transform(f_val)
f_test_scaled  = scaler.transform(f_test)

# ========================
# Class Weights
# ========================
class_weights = dict(enumerate(
    compute_class_weight('balanced', classes=np.unique(y_train_enc), y=y_train_enc)
))

# ========================
# Data Augmentation
# ========================
augment = tf.keras.Sequential([
    layers.RandomFlip("horizontal_and_vertical"),
    layers.RandomRotation(0.25),
    layers.RandomZoom(0.2),
    layers.RandomContrast(0.2),
    layers.RandomTranslation(0.1, 0.1)
])

# ========================
# Build CNN + ViT Model
# ========================
def build_model(num_classes):
    img_in  = layers.Input((IMG_SIZE, IMG_SIZE, 3))
    feat_in = layers.Input((4,))
    x = augment(img_in)

    # CNN backbone
    base = tf.keras.applications.ResNet50(include_top=False, weights='imagenet', input_tensor=x)
    for layer in base.layers[:-20]:
        layer.trainable = False
    cnn = layers.GlobalAveragePooling2D()(base.output)
    cnn = layers.Dense(64, activation='relu', kernel_regularizer=regularizers.l2(1e-4))(cnn)
    cnn = layers.Dropout(0.5)(cnn)

    # ViT branch
    p = layers.Conv2D(64, 16, strides=16)(x)
    p = layers.Reshape((-1, 64))(p)
    for _ in range(3):
        att = layers.MultiHeadAttention(num_heads=4, key_dim=64)(p, p)
        p = layers.LayerNormalization()(p + att)
        ff = layers.Dense(64, activation='relu')(p)
        p = layers.LayerNormalization()(p + ff)
    vit = layers.GlobalAveragePooling1D()(p)

    # Fusion
    m = layers.Concatenate()([cnn, vit, feat_in])
    m = layers.Dense(256, activation='relu', kernel_regularizer=regularizers.l2(1e-4))(m)
    m = layers.BatchNormalization()(m)
    m = layers.Dropout(0.5)(m)

    out = layers.Dense(num_classes, activation='softmax')(m)
    model = models.Model([img_in, feat_in], out)
    model.compile(
        optimizer=tf.keras.optimizers.Adam(1e-4),
        loss=tf.keras.losses.CategoricalCrossentropy(label_smoothing=0.1),
        metrics=['accuracy']
    )
    return model

model = build_model(len(np.unique(y_train_enc)))
model.summary()

# ========================
# Callbacks
# ========================
early_stop = tf.keras.callbacks.EarlyStopping(
    monitor='val_loss', patience=5, restore_best_weights=True
)

# ========================
# Train Model
# ========================
history = model.fit(
    [X_train, f_train_scaled], tf.keras.utils.to_categorical(y_train_enc),
    validation_data=([X_val, f_val_scaled], tf.keras.utils.to_categorical(y_val_enc)),
    epochs=EPOCHS,
    batch_size=BATCH_SIZE,
    class_weight=class_weights,
    callbacks=[early_stop],
    shuffle=True
)

# ========================
# Evaluate
# ========================
pred = np.argmax(model.predict([X_test, f_test_scaled]), axis=1)
print("\nTEST ACCURACY:", accuracy_score(y_test_enc, pred))
print(classification_report(y_test_enc, pred, target_names=le.classes_))

# ========================
# Plot Training Curves
# ========================
plt.figure(figsize=(12,5))
plt.subplot(1,2,1)
plt.plot(history.history['accuracy'], label='Train')
plt.plot(history.history['val_accuracy'], label='Val')
plt.title("Accuracy")
plt.legend()
plt.grid(True)

plt.subplot(1,2,2)
plt.plot(history.history['loss'], label='Train')
plt.plot(history.history['val_loss'], label='Val')
plt.title("Loss")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# ========================
# Save Model
# ========================
model.save("/content/drive/MyDrive/testprediction/cnn_vit_rbc_final.keras")
pickle.dump(le, open("/content/drive/MyDrive/testprediction/label_encoder.pkl", "wb"))
pickle.dump(scaler, open("/content/drive/MyDrive/testprediction/scaler.pkl", "wb"))
print("\n✅ FINAL MODEL SAVED SUCCESSFULLY")
```
